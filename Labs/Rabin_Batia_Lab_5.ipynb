```python
#if else statements
#if statements do something based on a condition

issue = "fever"

if issue == "headache":
  #boolean condition, expression, == determines if something is true
  print("Take an advil.")
  #if it is true than it will be printed
elif issue == "stomach ache":
  print("Take some Pepto Bismol.")
elif issue == "tired":
  print("Take a nap.")
else:
  print("Go to a doctor.")

#elif statements are read after if statements, starts with if, if true thats what it will do, if false it will continue down the elifs until it finds something that is true
```

```python
#can combine conditions by using and or or

age = 34

if age == 0 or age == 1:
  print("Infant")
elif age >= 2 and age <= 4:
  print("Toddler")
elif age >= 5 and age <= 11:
  print("Child")
elif age >= 12 and age <= 18:
  print("Teenager")
elif age >= 19 and age <= 65:
  print("Adult")
else:
  print("Wisened with age.")
```

```python
#Exercise 4

family = ["Bob", "Jeannie", "Tracy", "Samuel", "Sonya"]
#creating a list that has the names of a family

if len(family) >= 5:
  #if the number of people in the family list is 5 or more print this statement, checks the length of the list
  print("A large family.")
elif len(family) == 4:
  #if the number of people in the family list is 4 print this statement, checks the length of the list
  print("An average size family.")
else:
  #if the number of people in the family list is less than 4 print this statement
  print("A small family")
```

```python
#Question 4

grade = 85

if 100 >= grade >= 92:
  print("A")
elif 92 > grade >= 90:
  print("A-")
elif 90 > grade >= 87:
  print("B+")
elif 87 > grade >= 82:
  print("B")
elif 82 > grade >= 80:
  print("B-")
elif 80 > grade >= 77:
  print("C+")
elif 77 > grade >= 72:
  print("C")
elif 72 > grade >= 70:
  print("C-")
elif 70 > grade >= 60:
  print("D")
else:
  print("F")
```
