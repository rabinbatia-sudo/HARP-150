```python
#return statements
#used instead of print
#functions will stop running once a return statement is executed

def lower_number():
  num1 = int(input("Give me a number: "))
  num2 = int(input("Give me another number: "))

  if num1 < num2:
    return num1
  elif num2 < num1:
    return num2
  else:
    print("They are equal.")

low_number = lower_number()
print(low_number)
```

```python
#functions can return true or false
#functions can take multiple parameters
#they are called mandatory parameters, they are positional, first and second depend on how you order them

def addition(a,b):
  return a + b

print(addition(3, 6))
print(addition(4, 5))
print(addition(10, 15))

#there are also optional parameters
def math(a, b, operation = ""):
  if a > 0 and b > 0:
    if operation == "addition":
      return a + b
    elif operation == "subtraction":
      return a - b
    elif operation == "multiplication":
      return a * b
    else:
      return a, b

print(math(4, 6, "addition"))
print(math(4, 6, "multiplication"))
print(math(4, 6))

#if there are multiple optional parameters, we can choose to use one of them specifically by defining it
#for example, operation = "addition"
```

```python
#Exercise 4

def nums_greater_than_ten(numbers):
  over_ten = []
  for number in numbers:
    if number > 10:
      over_ten.append(number)
  return over_ten

nums = [1, 2, 10, 12, 7, 20, 50]
print(nums_greater_than_ten(nums))
```

```python
#Question 3

def money_in_week():
  hourly_wage = float(input("What is your hourly wage? "))
  hours_worked = int(input("How many hours do you work in a week? "))
  return hourly_wage * hours_worked

def money_for_weeks():
  weekly_wage = money_in_week()
  weeks = int(input("How many weeks in a year do you work? "))
  return weeks * weekly_wage

print(money_for_weeks())
```

```python
#Question 5

def money_per_order(order_amount, miles):
  gas_amount = miles * 3.60
  return order_amount - gas_amount + (order_amount * 0.15)

print(money_per_order(20, 3))
```

```python
#Question 6

#needs input from user
#need to randomly pick rock, paper or scissors by the program
#need to determine who wins, depending on what the randomized answer is

def rock_paper_scissors():
  your_pick = input("Choose rock, paper or scissors. ").lower()
  import random
  x = random.randint(1, 4)
  computer_pick = ""

  if x == 1:
    computer_pick = "rock"
  elif x == 2:
    computer_pick = "paper"
  elif x == 3:
    computer_pick = "scissors"

  if your_pick == "rock" and computer_pick == "paper":
    print("You lose.")
  elif your_pick == "rock" and computer_pick == "scissors":
    print("You win!")
  elif your_pick == "rock" and computer_pick == "rock":
    print("You tie.")
  elif your_pick == "paper" and computer_pick == "scissors":
    print("You lose.")
  elif your_pick == "paper" and computer_pick == "rock":
    print("You win!")
  elif your_pick == "paper" and computer_pick == "paper":
    print("You tie.")
  elif your_pick == "scissors" and computer_pick == "rock":
    print("You lose.")
  elif your_pick == "scissors" and computer_pick == "paper":
    print("You win!")
  elif your_pick == "scissors" and computer_pick == "scissors":
    print("You tie.")

rock_paper_scissors()
```
