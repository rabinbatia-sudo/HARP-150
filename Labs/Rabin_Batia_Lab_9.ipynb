```python
#dog class
class Dog:
  def __init__(self, name, age, color): #constructor
    self.name = name
    self.age = age
    self.color = color
    #always start a class with __init__ , short for initialize
    #assigning attributes and properties of the class
    #always have self, should always be the first parameter
  def greeting(self):
    print("Woof! My name is", self.name + " . I am", self.age, "years old. Do you like my", self.color, "fur?")
  def dog_years(self):
    """ Computes a dog's age into dog years. """
    #doc strings will show up when using the help function, have to be indented, used to explain methods
    print(self.age * 7)
  #Exercise 1
  def bark(self):
    print("Woof! Woof! Woof!")

d1 = Dog("Toby", 3, "yellow") #dog object, instance variable
d1.greeting()
d1.dog_years()
d1.bark()
```

```python
#Exercise 2

class store:
  def __init__(self, name, store_type, location):
    self.name = name
    self.store_type = store_type
    self.location = location
  def description(self):
    print(self.name, "is located at", self.location, "and it is a", self.store_type + ".")

s1 = store("Keyfoods", "grocery store", "Riverdale Ave")
s1.description()
s2 = store("Corner Cafe", "restaurant", "Johnson Ave")
s2.description()
```

```python
#Question 1

class BankAccount: #creating a class called BankAccount
    def __init__(self, balance): #constructor method, initializing, assigning attributes and properties of the class
        self.balance=balance #assigning an attribute to a specific instance of an object

    def get_balance(self): #creating a new method called get_balance
        return self.balance #returns the balance number

    def deposit(self,amt=0): #creating a new method called deposit and a new variable called amt
        self.balance+=amt #adding the amt number to the balance number

    def withdraw(self,amt=0): #creating a new method called withdraw
        self.balance-=amt #subtracting the amt number from the balance number

b1 = BankAccount(230) #creating a BankAccount object or instance variable with 230 as the balance
b1.deposit(100) #calling the deposit function and depositing 100
b1.withdraw(32) #calling the withdraw function and withdrawing 32
b1.get_balance() #calling the get_balance function to get the total balance after depositing and withdrawing

b2 = BankAccount(4450) #creating another BankAccount object with 4450 as the balance
b2.withdraw(2334) #calling the withdraw function and withdrawing 2334
b2.get_balance() #calling the get_balance function to get the total balance after withdrawing
```

```python
#Question 2

class Cat:
  def __init__(self, name, age, breed):
    self.name = name
    self.age = age
    self.breed = breed
  def description(self):
    print("The cat named", self.name, "is", self.age, "years old and is a", self.breed + ".")
  def cat_years(self):
    if self.age == 1:
      print(self.name, "is 15 in human years.")
    elif self.age == 2:
      print(self.name, "is 24 in human years.")
    else:
      human_years = ((self.age - 2) * 4) + 24
      print(self.name, "is", human_years, "in human years.")
  def pet(self):
    print("meow")

c1 = Cat("button", 5, "british shorthair")
c1.description()
c1.cat_years()
c1.pet()
```

```python
#Question 3

class odd_or_even:
  def __init__(self, num):
    self.num = num
  def odd_even(self):
    if self.num % 2 == 0:
      print("even")
    else:
      print("odd")

oe1 = odd_or_even(10)
oe1.odd_even()
```

```python
#Question 4

class BU_profile:
  def __init__(self, username, credits, class_standing):
    self.username = username
    self.credits = credits
    self.class_standing = class_standing
  def log_in(self):
    print(self.username)
    password = input("Please type in your password: ")
    print("You have logged into your Binghamton University account!")
  def credits_left(self):
    print("You have", 126 - self.credits, "credits left to take before you can graduate.")
  def classes(self):
    if self.class_standing == "freshman":
      print("You can take 100 and 200 level classes.")
    elif self.class_standing == "sophmore":
      print("You can take 100, 200 and 300 level classes.")
    elif self.class_standing == "junior":
      print("You can take 200, 300 and 400 level classes.")
    else:
      print("You can take 300 and 400 level classes.")

b1 = BU_profile("brabin1", 22, "freshman")
b1.log_in()
b1.credits_left()
b1.classes()
```
