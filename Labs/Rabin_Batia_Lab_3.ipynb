```python
#Lists, Tuples, and Sets
#lists are mutable (changeable), ordered, can have duplicates, any data type, nested (list of lists), use brackets, usually used because it's mutable
num_list = [1, 2, 3, 4, 5]
list_nums = [5, 4, 3, 2, 1]
num_list == list_nums
#gives false
#the == is for equivalency, it's a boolean
```

```python
#indexing lists
color_list = ["red", "green", "blue", "yellow", "purple", "black"]
print(color_list[0])
#index numbers start with zero
color_list.index("green")
#gives the number of the index
#if you try to get an index that's not in the list you get an error, index doesn't exist
#can split lists
print(color_list[2:5])
print(color_list[:3])
#this can also be done with strings
name = "Selena Gomez"
print(name[4:6])
#takes from the start number to the number before the end number
#can also find the length of lists
list_length = len(color_list)
print(list_length)
#can also do math things with lists of integers, sum adds them all together, min finds lowest number, max finds highest
my_nums = [34, 53, 22, 64, 82, 11]
print(sum(my_nums))
print(max(my_nums))
print(min(my_nums))
#can have the name of a variable at the bottom and it will print the variable, not suggested to use, but is essentially the same result
```

```python
#adding to a list, add to the end
color_list = ["red", "green", "blue", "yellow", "purple", "black"]
color_list.append("grey")
#this will add it to the end
color_list.insert(3, "pink")
#this will add it at index 3
print(color_list)
#sometimes if you run the same cell a bunch of times that has append, you can get the thing that is getting appended a bunch of times at the end
```

```python
#two lists can be added together with the extend method
colors_1 = ["red", "green", "blue"]
colors_2 = ["yellow", "orange", "purple"]
colors_1.extend(colors_2)
print(colors_1)
```

```python
#can remove from a list
colors_list = ["red", "green", "blue", "yellow", "purple", "black"]
colors_list.remove("yellow")
#will remove the first instance of it not all of it in the list
#can remove a specific index with pop()
colors_list.pop(0)
colors_list.pop() #will take the last index
```

```python
#set, unordered, do not contain duplicates, use squiggly brackets, can't be indexed, uses less memory, used to get rid of duplicates
#can convert lists to sets
#set(colors_list)
nums = [1, 2, 2, 2, 3, 3, 3, 4, 4, 4]
nums_set = set(nums)
print(nums_set)
#can find the difference between two groups
#one_set.difference(another_set), finds the difference
#can combine two sets
#one_set.union(another_set), joins them together in no particular order
```
