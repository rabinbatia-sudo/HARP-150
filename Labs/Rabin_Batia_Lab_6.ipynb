```python
#for loops allows us to iterate, allows to repeat
#structure of a for loop
#for key (index) in abbreviations (range):
# print(abbreviations[key])

sequence = range(4,8)
for item in sequence:
  print(item)
#range can get it in the for loop directly as well
#for item in range (4,8)

for item in range(4,8):
  if item != 6:
    print(item)

#print from the start to one before the end

for number in range(10):
  print(number)
  #will start at 0 and go to 10

for number in range(0, 39, 3):
  print(number, end = " ")
print()
#will start at 0 end at 39 and increase by three each time
```

```python
#can store data in a list, move data from one list to another with a for loop
#while loop will keep going until the condition is met, if it never meets the condition it will run forever

x = 1
total = 0

while x <= 10:
  total += x
  x += 1
print(total)

#print debugging, print statements to figure out whats going on in the code at any point in time
#while loops can support multiple conditions
#make sure your while loop has a stop condition
#can use else clause with while loops
```

```python
for i in range(0, 10):
  if i == 5:
    continue
    #continue skips an iteration
  elif i == 7:
    break
    #break stops a while loop at a specific point
  print(i)
```

```python
#Question 2

number = 1

while 1 <= number < 10:
  if number % 2 == 0:
    print("Even")
  elif number % 2 == 1:
    print("Odd")
  number += 1
```

```python
#Question 3

number = 1
even_list = []
odd_list = []

while 1 <= number < 10:
  if number % 2 == 0:
    even_list.append(number)
  elif number % 2 == 1:
    odd_list.append(number)
  number += 1
print(even_list)
print(odd_list)
```

```python
#Question 4

courses = {"Monday": 20, "Tuesday": 25, "Wednesday": 13, "Thursday": 5, "Friday": 25}

for course in courses:
  if courses[course] == 25:
    print("The", course, "course is full.")
  elif 20 <= courses[course] <= 24:
    print("The", course, "course is almost full.")
  elif courses[course] < 20:
    print("There are plenty of seats left in the", course, "course.")
```

```python
#Question 6

shopping_list = {"chocolate milk": 5,
                 "pizza": 8,
                 "fries": 10,
                 "milk": 7,
                 "orange juice": 9,
                 "cheese": 9,
                 "pasta": 6}
money = 50

for item in shopping_list:
  if money - shopping_list[item] >= 0:
    money -= shopping_list[item]
print(money)
```
