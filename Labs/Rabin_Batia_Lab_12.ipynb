```python
import pandas as pd
data = pd.read_csv("Real_Estate_Sales_2001-2020_GL.csv", low_memory = False)
#saying low_memory = False will help to run the code when there is a huge data set
#error from running out of memory, very large data set
#how much have house prices have increased on average, between 2010 and 2020 in Conneticut?
data
```

```python
#Exercise 2

dict_homes_year = {}
new = data["List Year"]

for i in new:
  if i not in dict_homes_year:
    dict_homes_year[i] = 1
  else:
    dict_homes_year[i] += 1

dict_homes_year
```

```python
import seaborn as sns

cols = data.columns[:30] #first 30 columns
colors = ["#000099", "#ffff00"]
#specify the colors - yellow is missing, blue is not
sns.heatmap(data[cols].isnull(), cmap = sns.color_palette(colors))
```

```python
#can easily delete columns with .drop()
data2 = data.copy()
#creating a copy first to keep an intact version

data2.drop(labels = ["OPM remarks", "Location", "Non Use Code", "Assessor Remarks"], axis = 1, inplace = True)
```

```python
#lets get rid of rows to get the range in values that we want
data2 = data2.drop(data2[(data2["List Year"] < 2010)].index, inplace = False)
data2 = data2.drop(data2[(data2["List Year"] > 2020)].index, inplace = False)
#if you do inplace = true you don't have to make it equal to data2 if you make inplace = false you do
data2
```

```python
#have to address rows with null values
#could use .dropna() to delete all the rows with NaN values but this will ruin our data set
#so instead we can use .fillna()
data2["Property Type"] = data2["Property Type"].fillna("Not Specified")
#grab the column and make it equal to the new column we created to make it work
data2
#property type should be completely blue on the heat map
```

```python
avg_sale = data2.groupby("List Year")[["Sale Amount"]].mean().round()
avg_sale
```

```python
#Exercise 3

avg = data2.groupby("Town")[["Assessed Value"]].mean().round()
avg.sort_values("Assessed Value", ascending = False)
```

```python
import matplotlib.pyplot as plt

avg_sale.plot(kind = "bar", y = "Sale Amount", color = "Purple")
plt.show()
```

```python
#Question 1

import pandas as pd
data3 = pd.read_csv("Real_Estate_Sales_2001-2020_GL.csv", low_memory = False)
data3.drop(labels = ["OPM remarks", "Location", "Non Use Code", "Assessor Remarks"], axis = 1, inplace = True)
data3
```

```python
#Question 2

data3 = data3.rename(columns={"Serial Number": "serial_number",
                                          "List Year": "list_year",
                                          "Date Recorded": "date_recorded",
                                          "Town": "town",
                                          "Address": "address",
                                          "Assessed Value": "assessed_value",
                                          "Sale Amount": "sale_amount",
                                          "Sales Ratio": "sales_ratio",
                                          "Property Type": "property_type",
                                          "Residential Type": "residential_type"})
```

```python
#Question 3

data3 = data3.drop(data3[(data3["list_year"] < 2001)].index, inplace = False)
data3 = data3.drop(data3[(data3["list_year"] > 2009)].index, inplace = False)
```

```python
#Question 4

data4 = data3.copy()
data4 = data4.dropna()
data4
```

```python
#Question 5

data3["property_type"] = data3["property_type"].fillna("Not Specified")
data3["residential_type"] = data3["residential_type"].fillna("Not Specified")
data3["sales_ratio"] = data3["sales_ratio"].fillna("Not Specified")
data3
```

```python
#Question 6

avg = data3.groupby("list_year")[["sale_amount"]].mean().round()
avg.sort_values("list_year", ascending = False)
```

```python
#Question 7

import numpy as np
import matplotlib.pyplot as plt

x = np.array([2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009])
y = np.array([246235, 296357, 327218, 380297, 364030, 510645, 368883, 435899, 336341])

font = {'family':'helvetica','size':15}

plt.title("Average Sale Amount Per List Year", fontdict = font)
plt.xlabel("List Year", fontdict = font)
plt.ylabel("Average Sale Amount", fontdict = font)

plt.plot(x, y)
plt.show()
```
