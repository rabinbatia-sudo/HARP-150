```python
#make charts based on data we input ourselves

import matplotlib.pyplot as plt
#same as before just a different way

labels = "Freshmen", "Sophmore", "Junior", "Senior"
numbers = [15, 30, 45, 10]
#percent of the pie chart each takes up
explode = (0, 0, 0.1, 0)
#takes a part of the pie chart off of the chart and seperates it
#numbers and explode go in order of the labels

fig1, ax1 = plt.subplots()
#sets the figure name an axis
ax1.pie(numbers, explode = explode, labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
#gives the paramaters of the axis, use the numbers from before, make explode equal explode, labels equal labels
#autopct allows a pct sign, shadow adds shadow to the graph, makes the start angle 90
#everything but wedge sizes are optional parameters
ax1.axis('equal')
#sets an equal aspect ratio

plt.show()
```

```python
dict_category = {}

new = permits["Category"]

for i in new:
  if i not in dict_category:
    dict_category[i] = 1
  else:
    dict_category[i] += 1

print(dict_category)
#showing the frequency of each value in the category column
```

```python
keys_category = list(dict_category.keys())
#all the keys of the dictionary

values_category = list(dict_category.values())
#all the values of the dictionary

print(keys_category)
print(values_category)
```

```python
plt.pie(values_category, labels = keys_category, autopct = "%1.1f%%")
plt.title("Distribution of Permits")
plt.show()
```
