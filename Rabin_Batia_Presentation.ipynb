```python
#for loops iterate through a list
#every for loop has an index and a range

for item in range(6,10):
  print(item / 2)

sequence = range(10)

for item in sequence:
  if item % 2 == 0:
    print(item)

#the range is exclusive, it doesn't include the last number

for item in range(0, 1000, 100):
  print(item)

#can also do nested for loops

for i in range(1, 11):
  for j in range(1, 11):
    print(i * j, end = " ")
  print()
```

```python
#can iterate through a dictionary or list

monsters = ["dragon", "goblin", "orc", "kraken"]

for monster in monsters:
  print("Would you like to fight a", monster, "?")
```

```python
#let's look at an example of something similar to the games we saw last class

monsters = { "dragon": 30, "goblin": 5, "orc": 15, "kraken": 50}
health = 20
monsters_killed = []
monsters_killed_by =[]

for monster in monsters:
  if monsters[monster] <= health:
    monsters_killed.append(monster)
  elif monsters[monster] > health:
    monsters_killed_by.append(monster)
print("You killed", monsters_killed)
print("You were killed by", monsters_killed_by)
```

```python
#while loops run as long as its condition is true, if it has no limit it can run forver which is called an infinite loop

x = 1
total = 0

while x <= 10:
  total += x
  x += 1

print(total)

w = 10
total = 0

while w >= 0:
  total += w
  w -= 1

print(total)

y = 10
z = 20

while y <= 50 and z <= 70:
  y += 10
  z += 10
  print(y, z)

#if you create an infinte loop press the stop button in JN
```

```python
#else clauses can also be used with while loops

x = 0

while x <= 10:
  print(x)
  x += 1
else:
  print("The loop is finsihed.")
```

```python
#breaks and continues
#break stops the code completely when it reaches that line
#continue skips an iteration
#let's use question 5 from the homework as an example

name = "Queen Elizabeth II"

for item in name:
  if item == "i":
    break
  print(item)
```

```python
name = "Queen Elizabeth II"

for item in name:
  if item == "i":
    continue
  print(item)
```
