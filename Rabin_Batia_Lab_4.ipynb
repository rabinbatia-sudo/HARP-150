```python
#Dictionaries hold both keys and values
#They us curly brackets
#Cannot have the same key twice, store things in key-value pairs
name_of_dictionary = {"Key1": "Value1",
                      "key2": "Value2",
                      "Key3:": "Value3",
                      "Key4": "Value4"}
#Dictionaries are mutable, ordered, and don't contain duplicates
dict_names = {"cousin1": "Alex",
             "cousin2": "Phil",
             "cousin3": "Jill"}
print(dict_names["cousin2"])
#no duplicate keys are allowed, the most recent one under teh key name will be the one accessed
#keys can be reassigned
dict_names["cousin2"] = "Tara"
#keys can also be added, they will be added at the end
dict_names["cousin4"] = "Phil"
print(dict_names)
#can check if something is in the dictionary we can use the in value, will return true or false
"cousin3" in dict_names
#length function can be used to show the size of the dictionary
len(dict_names)
#can store lists and dictionaries in a dicionary, nested dictionaries
#used alot to count how often something shows up
```
