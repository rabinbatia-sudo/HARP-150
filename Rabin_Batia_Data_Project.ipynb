```python
class list_avg:
  def __init__(self, list1, list2):
    self.list1 = list1
    self.list2 = list2

  def to_be_dict(self):
    self.new_dict = df.groupby(self.list1)[self.list2].apply(list).to_dict()
    #https://stackoverflow.com/questions/29876184/groupby-results-to-dictionary-of-lists
    return self.new_dict
    #turning the two columns into lists and grouping them into a dictionary, list1 is the keys, list2 the values

  def avg_dict(self):
    for i in self.new_dict:
      self.key_list = self.new_dict.get(i)
      count = 0
      sum = 0
      for j in self.key_list:
        sum += j
        count += 1
      self.new_dict[i] = round(sum/count, 2)
      #https://www.w3schools.com/python/ref_func_round.asp
    self.averages = self.new_dict
    return self.averages
    #finding the mean of the values of new_dict and rounding them for each country

  def valuess(self):
    return self.averages.values()

  def keyss(self):
    return self.averages.keys()


avg_dbr = list_avg("country", "domestic_beer_0.5liter_restaurants")
avg_dbr.to_be_dict()
avg_dbr.avg_dict()
dbr_values = avg_dbr.valuess()

avg_ibr = list_avg("country", "imported_beer_0.33liter_restaurants")
avg_ibr.to_be_dict()
avg_ibr.avg_dict()
ibr_values = avg_ibr.valuess()

avg_dbm = list_avg("country", "domestic_beer_0.5liter_market")
avg_dbm.to_be_dict()
avg_dbm.avg_dict()
dbm_values = avg_dbm.valuess()

avg_ibm = list_avg("country", "imported_beer_0.33liter_market")
avg_ibm.to_be_dict()
avg_ibm.avg_dict()
ibm_values = avg_ibm.valuess()
```

```python
class dict_avg(list_avg):
  #child class of list_avg
  def __init__(self, list1, list2, list3, list4, list5, list6, list7, list8, list9, list10, list11, list12, list13, list14, list15):
    super().__init__(list1, list2)
    self.list3 = list3
    self.list4 = list4
    self.list5 = list5
    self.list6 = list6
    self.list7 = list7
    self.list8 = list8
    self.list9 = list9
    self.list10 = list10
    self.list11 = list11
    self.list12 = list12
    self.list13 = list13
    self.list14 = list14
    self.list15 = list15

  def to_avg_dicts(self):
    self.l1 = list_avg(self.list1, self.list2)
    self.l2 = list_avg(self.list1, self.list3)
    self.l3 = list_avg(self.list1, self.list4)
    self.l4 = list_avg(self.list1, self.list5)
    self.l5 = list_avg(self.list1, self.list6)
    self.l6 = list_avg(self.list1, self.list7)
    self.l7 = list_avg(self.list1, self.list8)
    self.l8 = list_avg(self.list1, self.list9)
    self.l9 = list_avg(self.list1, self.list10)
    self.l10 = list_avg(self.list1, self.list11)
    self.l11 = list_avg(self.list1, self.list12)
    self.l12 = list_avg(self.list1, self.list13)
    self.l13 = list_avg(self.list1, self.list14)
    self.l14 = list_avg(self.list1, self.list15)
    #creating 14 list_avg objects

    self.l1.to_be_dict()
    self.l2.to_be_dict()
    self.l3.to_be_dict()
    self.l4.to_be_dict()
    self.l5.to_be_dict()
    self.l6.to_be_dict()
    self.l7.to_be_dict()
    self.l8.to_be_dict()
    self.l9.to_be_dict()
    self.l10.to_be_dict()
    self.l11.to_be_dict()
    self.l12.to_be_dict()
    self.l13.to_be_dict()
    self.l14.to_be_dict()
    #turning the list_avg objects into dictionaries

    self.ad1 = self.l1.avg_dict()
    self.ad2 = self.l2.avg_dict()
    self.ad3 = self.l3.avg_dict()
    self.ad4 = self.l4.avg_dict()
    self.ad5 = self.l5.avg_dict()
    self.ad6 = self.l6.avg_dict()
    self.ad7 = self.l7.avg_dict()
    self.ad8 = self.l8.avg_dict()
    self.ad9 = self.l9.avg_dict()
    self.ad10 = self.l10.avg_dict()
    self.ad11 = self.l11.avg_dict()
    self.ad12 = self.l12.avg_dict()
    self.ad13 = self.l13.avg_dict()
    self.ad14 = self.l14.avg_dict()
    #finding the average of the values for each dictionary

  def consolidate(self):
    self.dicts = [self.ad1, self.ad2, self.ad3, self.ad4, self.ad5, self.ad6, self.ad7, self.ad8, self.ad9, self.ad10, self.ad11, self.ad12, self.ad13, self.ad14]
    self.con_dict = {}
    for h in self.ad1.keys():
      self.con_dict.update({h: []})
    for i in self.dicts:
      for j in i:
        if j in self.con_dict:
          self.value = self.con_dict[j]
          self.value.append(i.get(j))
    return self.con_dict
    #consolidating and combining all of the averaged dictionaries together

  def most_expensive(self):
    for i in self.con_dict:
      max_num = max(self.con_dict.get(i))
      #https://www.geeksforgeeks.org/python-program-to-find-largest-number-in-a-list/
      #finding the largest number in a list
      if self.con_dict.get(i).index(max_num) == 0:
        self.con_dict[i] = "milk_1liter"
      elif self.con_dict.get(i).index(max_num) == 1:
        self.con_dict[i] = "loaf_fresh_white_bread_500g"
      elif self.con_dict.get(i).index(max_num) == 2:
        self.con_dict[i] = "white_rice_1kg"
      elif self.con_dict.get(i).index(max_num) == 3:
        self.con_dict[i] = "dozen_eggs"
      elif self.con_dict.get(i).index(max_num) == 4:
        self.con_dict[i] = "local_cheese_1kg"
      elif self.con_dict.get(i).index(max_num) == 5:
        self.con_dict[i] = "chicken_fillets_1kg"
      elif self.con_dict.get(i).index(max_num) == 6:
        self.con_dict[i] = "beef_round_1kg"
      elif self.con_dict.get(i).index(max_num) == 7:
        self.con_dict[i] = "apples_1kg"
      elif self.con_dict.get(i).index(max_num) == 8:
        self.con_dict[i] = "banana_1kg"
      elif self.con_dict.get(i).index(max_num) == 9:
        self.con_dict[i] = "oranges_1kg"
      elif self.con_dict.get(i).index(max_num) == 10:
        self.con_dict[i] = "tomato_1kg"
      elif self.con_dict.get(i).index(max_num) == 11:
        self.con_dict[i] = "potato_1kg"
      elif self.con_dict.get(i).index(max_num) == 12:
        self.con_dict[i] = "onion_1kg"
      elif self.con_dict.get(i).index(max_num) == 13:
        self.con_dict[i] = "lettuce_1head"
      #depending on what the index of the largest number is determines which food product is the most expensive in that country
      #since the numbers were all added in order, the index corelates with what column/list it came from

    self.expenses = self.con_dict
    return self.expenses

  def percentages(self):
    self.amounts = {}
    for i in self.expenses.values():
      if i not in self.amounts:
        self.amounts[i] = 1
      else:
        self.amounts[i] += 1
    return self.amounts
    #counts how many of each food product is in the list


da1 = dict_avg("country", "milk_1liter", "loaf_fresh_white_bread_500g", "white_rice_1kg", "dozen_eggs", "local_cheese_1kg",
               "chicken_fillets_1kg", "beef_round_1kg", "apples_1kg", "banana_1kg", "oranges_1kg", "tomato_1kg",
               "potato_1kg", "onion_1kg", "lettuce_1head")
da1.to_avg_dicts()
da1.consolidate()
da1.most_expensive()
da1p = da1.percentages()
print(da1p)
```

```python
class expensive_apartments(list_avg):
  #child class of list_avg
  def __init__(self, list1, list2, list3, list4):
    super().__init__(list1, list2)
    self.list2list = list(df[list2])
    #https://medium.com/geekculture/python-list-comprehension-with-pandas-6514557babc6
    self.list3 = list3
    self.list4 = list4

  def hundred_most(self):
    self.dict1 = df.groupby(self.list2)[self.list1].apply(list).to_dict()
    self.dict2 = {}
    self.dict3 = {}
    self.final_list = []

    for i in range(0, 100):
        self.max = 0
        for j in range(len(self.list2list)):
            if self.list2list[j] > self.max:
                self.max = self.list2list[j]
        self.list2list.remove(self.max)
        self.final_list.append(self.max)

    #https://www.geeksforgeeks.org/python-program-to-find-n-largest-elements-from-a-list/
    #lines 13 - 21 were taken from the website above

    for i in self.final_list:
      for j in self.dict1:
        if i not in self.dict2 and j == i:
          self.dict2[i] = self.dict1.get(j)

    return self.dict2

  def the_keys(self):
    self.dict2_values = list(self.dict2.values())
    #https://stackoverflow.com/questions/33674033/python-how-to-convert-a-dictionary-into-a-subscriptable-array
    #turns dictionary values into a list, very important so later on these values can be used as keys in a dictionary
    self.all_keys = []
    self.count = 0

    for i in self.dict2_values:
      if 2 >= len(i) > 1:
        self.all_keys.append("Lubumbashi")
        self.all_keys.append("Palm Bay")
        self.count += 1
      elif 5 >= len(i) > 4:
        self.all_keys.append("Lusaka")
        self.all_keys.append("Oceanside")
        self.all_keys.append("Malden")
        self.all_keys.append("Newport Beach")
        self.all_keys.append("Encinitas")
        self.count += 1
      #the if and elif are aimed at two lists within the list of values to take them out of the list
      else:
        self.all_keys.append(self.dict2_values[self.count][0])
        #https://stackoverflow.com/questions/18449360/access-item-in-a-list-of-lists
        #this turned the values from list objects back into strings
        self.count += 1

    return self.all_keys

  def specifics1(self):
    self.l1 = list_avg(self.list1, self.list3)
    self.d1 = self.l1.to_be_dict()
    self.d1values = {}

    for i in self.all_keys:
      if i not in self.d1values:
        self.d1values[i] = max(self.d1.get(i))
        #https://www.geeksforgeeks.org/python-program-to-find-largest-number-in-a-list/
        #some of the keys had more than one value, this gave the largest value
        #this also helped to turn the values from list objects into integers

    return self.d1values

  def specifics2(self):
    self.l2 = list_avg(self.list1, self.list4)
    self.d2 = self.l2.to_be_dict()
    self.d2values = {}

    for i in self.all_keys:
      if i not in self.d2values:
        self.d2values[i] = max(self.d2.get(i))

    return self.d2values

a1 = expensive_apartments("city", "price_per_sqm_buy_apartment_city_center", "1bedroom_apartment_city_center", "3bedroom_apartment_city_center")
h = a1.hundred_most()
k = a1.the_keys()
v1 = a1.specifics1()
v2 = a1.specifics2()
```
