```python
#Exercise 2

class Employee:
  def __init__(self, first, last, salary):
    self.first = first
    self.last = last
    self.salary = salary

  def fullname(self):
    print(f"{self.first} {self.last}")

  def monthlySalary(self):
    return self.salary/12

class Professor(Employee):
  def __init__(self, first, last, salary, department, classes):
    super().__init__(first, last, salary)
    self.department = department
    self.classes = classes

  def introduction(self):
    print(f"Professor {self.first} {self.last} is in the {self.department} department and is teaching {self.classes} classes.")

p1 = Professor("Casey", "Doyle", 60000, "philosophy", 3)
p1.fullname()
p1.introduction()
```

```python
#Question 1

class BinghamtonID:
  def __init__(self, first_name, last_name):
    self.first = first_name
    self.last = last_name
  def email(self):
    if len(self.last) >= 6:
      return self.first[0] + self.last[0:6]
    else:
      return self.first[0] + self.last

b1 = BinghamtonID("batia", "rabin")
b2 = BinghamtonID("gavriella", "lapin")
b3 = BinghamtonID("benjamin", "shalumov")

print(b1.email())
print(b2.email())
print(b3.email())
```

```python
#Question 2

class Animal:
  def __init__(self, weight, age, name):
    self.weight = weight
    self.age = age
    self.name = name

  def introduction(self):
    print(f"Hi, my name is {self.name} I'm {self.age} years old and I weigh {self.weight} pounds.")

  def name_length(self):
    if len(self.name) >= 8:
      print("I have a long name.")
    elif 4 <= len(self.name) < 7:
      print("I have a normal length name.")
    else:
      print("I have a short name.")

class Dog(Animal):
  def __init__(self, weight, age, name, breed, hair):
    super().__init__(weight, age, name)
    self.breed = breed
    self.hair = hair

  def dog_years(self):
    print(self.age * 7)

  def dog_type(self):
    print(f"I'm a {self.breed} dog and my hair is {self.hair}")

  def bark(self):
    print("Woof! Woof! Woof!")

class Cat(Animal):
  def __init__(self, weight, age, name, breed, color):
    super().__init__(weight, age, name)
    self.breed = breed
    self.color = color

  def cat_years(self):
    if self.age == 1:
      print(self.name, "is 15 in human years.")
    elif self.age == 2:
      print(self.name, "is 24 in human years.")
    else:
      human_years = ((self.age - 2) * 4) + 24
      print(self.name, "is", human_years, "in human years.")

  def cat_type(self):
    print(f"I'm a {self.breed} cat and my fur is {self.color}.")

c1 = Cat(20, 5, "button", "exotic shorthair", "cider")
c1.introduction()
c1.cat_years()
```

```python
#Question 3

class Student:
  def __init__(self, first_name, grade_dict, school_name, age):
    self.name = first_name
    self.grades = grade_dict
    self.school = school_name
    self.age = age

  def check_grade(self):
    for i in self.grades:
      if self.grades[i] < 70:
        print(f"You need to get some tutoring in {i}!")
      else:
        print(f"You're doing great in {i}!")
  def school_type(self):
    if self.age <= 5:
      print("You are in pre-school.")
    elif 5 < self.age <= 11:
      print("You are in elementary school.")
    elif 11 < self.age <= 14:
      print("You are in middle school.")
    elif 14 < self.age <= 18:
      print("You are in high school.")
    elif 18 < self.age <= 22:
      print("You are in college.")
    elif 22 < self.age <= 26:
      print("You are in graduate school.")

  def introduction(self):
    print(f"Hi, my name is {self.name}, I'm {self.age} years old, I go to {self.school} and my grades are {self.grades}.")

class Art_Student(Student):
  def __init__(self, first_name, grade_dict, school_name, age):
    super().__init__(first_name, grade_dict, school_name, age)
    self.good_art_students = ["Robbie", "Christine", "Elliott"]

  def good_art(self):
    if self.grades["art"] >= 90:
      self.students.append(self.name)
      return self.good_art_students

Student001 = Art_Student("Abdul", {"math": 85, "science": 60, "art": 90, "history": 34}, "Binghamton University", 21)

Student001.introduction()
Student001.check_grade()
Student001.school_type()
print(Student001.good_art())
```

```python
#Question 4

class Player:
  def __init__(self, name, hp):
    self.name = name
    self.hp = hp
    self.damage = 0
    self.inventory = []
    self.damage_monster = 0

  def calculate_health(self):
    return self.hp - self.damage

  def add_items(self, item):
    self.item = item
    self.inventory.append(self.item)

class Paladin(Player):
  def __init__(self, name, hp, strength, faith):
    super().__init__(name, hp)
    self.strength = strength
    self.faith = faith

  def DivineSmite(self):
    if self.hp <= 10:
      self.strength += 5
      self.damage_monster += 5
    else:
      self.strength -= 5
      self.damage += 5

  def ChannelDivinity(self):
    if self.faith >= 100:
      self.strength += 20
      self.hp += 10
    else:
      self.strength -= 20
      self.hp -= 10

class Mage(Player):
  def __init__(self, name, hp, mana, intelligence):
    super().__init__(name, hp)
    self.mana = mana
    self.intelligence = intelligence

  def ArcaneBolt(self):
    if 20 > self.mana >= 10:
      self.damage_monster += 5
    elif 30 > self.mana >= 20:
      self.damage_monster += 10
    elif 40 > self.mana >= 30:
      self.damage_monster += 15
    else:
      self.damage_monster += 20

  def ArcaneFocus(self):
    if 20 > self.intelligence >= 10:
      self.mana += 5
    elif 30 > self.intelligence >= 20:
      self.mana += 10
    elif 40 > self.intelligence >= 30:
      self.mana += 15
    else:
      self.mana += 20


p1 = Player("Sallow", 34)
print(p1.calculate_health())
p1.add_items("water")
print(p1.inventory)
```
