```python
#functions with turtle

def draw_triangle(side_length):
  for i in range(side_length):
    t.fd(side_length)
    t.rt(120)

draw_triangle(200)

def draw_rectangle(length, height):
  for i in range(0, 4):
    if i % 2 == 0:
      t.fd(length)
      t.rt(90)
    else:
      t.fd(height)
      t.rt(90)

draw_rectangle(200, 100)
```

```python
#Exercise 4

import turtle
from random import randint

t1 = turtle.Turtle()
t1.shape("turtle")
t1.color("purple")
t2 = t1.clone()
t2.color("red")
t3 = t1.clone()
t3.color("blue")

t1.penup()
t2.penup()
t3.penup()
t1.goto(50, 50)
t2.goto(-50, 50)
t3.goto(-50, -50)
t1.pendown()
t2.pendown()
t3.pendown()

t1_speed = randint(0, 3)
t2_speed = randint(0, 3)
t3_speed = randint(0, 3)

t1.speed(t1_speed)
t2.speed(t2_speed)
t3.speed(t3_speed)

t1.fd(300)
t2.fd(300)
t3.fd(300)

if t1_speed > t2_speed and t1_speed > t3_speed:
  print("The purple turtle won!")
elif t2_speed > t1_speed and t2_speed > t3_speed:
  print("The red turtle won!")
elif t3_speed > t1_speed and t3_speed > t2_speed:
  print("The blue turtle won!")
else:
  print("There was a tie.")
```

```python
#Homework

import turtle
from random import randint

t1 = turtle.Turtle()
t1.shape("turtle")
t1.color("purple")
t2 = t1.clone()
t2.color("green")
t1.speed(10)
t2.speed(10)
quadrant1 = int(input("What quadrant do you want to count for the purple turtle? "))
quadrant2 = int(input("What quadrant do you want to count for the green turtle? "))

def draw_squares(quad_t1, quad_t2):
  count1 = 0
  count2 = 0
  x = 0
  while x <= 20:
    for i in range(0, 4):
      t1.fd(10)
      t1.rt(90)
      t2.fd(10)
      t2.rt(90)
    t1x = randint(-300, 300)
    t1y = randint(-400, 400)
    t2x = randint(-300, 300)
    t2y = randint(-400, 400)
    t1.penup()
    t2.penup()
    t1.goto(t1x, t1y)
    t2.goto(t2x, t2y)
    t1.pendown()
    t2.pendown()
    if quad_t1 == 1:
      if 0 <= t1x <= 300 and 0 <= t1y <= 400:
        count1 += 1
    elif quad_t1 == 2:
      if -300 <= t1x <= 0 and 0 <= t1y <= 400:
        count1 += 1
    elif quad_t1 == 3:
      if -300 <= t1x <= 0 and -400 <= t1y <= 0:
        count1 += 1
    else:
      if 0 <= t1x <= 300 and -400 <= t1y <= 0:
        count1 += 1
    if quad_t2 == 1:
      if 0 <= t1x <= 300 and 0 <= t1y <= 400:
        count2 += 1
    elif quad_t2 == 2:
      if -300 <= t1x <= 0 and 0 <= t1y <= 400:
        count2 += 1
    elif quad_t2 == 3:
      if -300 <= t1x <= 0 and -400 <= t1y <= 0:
        count2 += 1
    else:
      if 0 <= t1x <= 300 and -400 <= t1y <= 0:
        count2 += 1
    x += 1
  if count1 > count2:
    print("The purple turtle wins!")
  elif count2 > count1:
    print("The green turtle wins!")
  else:
    print("There is a tie.")

draw_squares(quadrant1, quadrant2)
```
